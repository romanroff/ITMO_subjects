Задача 5. Паутинная модель рынка.

Для нахождения равновесной цены $ p $ в паутинной модели рынка, мы должны установить условие равновесия, при котором спрос равен предложению, т.е. $ d_t = s_t $. В равновесии цена не изменяется со временем, поэтому $ p_t = p_{t-1} = p $.

Исходные уравнения:
$$
d_t = a - b p_t
$$
$$
s_t = m + n p_{t-1}
$$

В равновесии:
$$
a - b p = m + n p
$$

Подставим значения коэффициентов для $ N = 4 $:
$$
a = 4, \quad b = \frac{4 + 2}{4} = \frac{6}{4} = \frac{3}{2}, \quad m = \frac{4}{2} = 2, \quad n = \frac{4}{4 + 3} = \frac{4}{7}
$$

Теперь подставим эти значения в уравнение равновесия:
$$
4 - \frac{3}{2} p = 2 + \frac{4}{7} p
$$

Перенесем все члены с $ p $ в одну сторону, а свободные члены в другую:
$$
4 - 2 = \frac{3}{2} p + \frac{4}{7} p
$$
$$
2 = \left( \frac{3}{2} + \frac{4}{7} \right) p
$$

Приведем коэффициенты при $ p $ к общему знаменателю:
$$
\frac{3}{2} = \frac{21}{14}, \quad \frac{4}{7} = \frac{8}{14}
$$
$$
\frac{21}{14} + \frac{8}{14} = \frac{29}{14}
$$

Теперь уравнение принимает вид:
$$
2 = \frac{29}{14} p
$$

Решим относительно $ p $:
$$
p = \frac{2 \times 14}{29} = \frac{28}{29}
$$

Таким образом, равновесная цена $ p $ равна $ \frac{28}{29} $.
