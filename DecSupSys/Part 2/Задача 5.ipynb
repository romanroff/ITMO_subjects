Задача 5. Паутинная модель рынка.

1. Динамика изменения цен:
   $$
   d_t = a - b p_t, \quad s_t = m + n p_{t-1}.
   $$
   В условиях рыночной динамики спрос равен предложению:
   $$
   a - b p_t = m + n p_{t-1}.
   $$

2. Перепишем разностное уравнение для $ p_t $:
   $$
   p_t = \frac{a - m}{b} - \frac{n}{b} p_{t-1}.
   $$

3. Подставим значения коэффициентов:
   $$
   a = 4, \quad b = \frac{3}{2}, \quad m = 2, \quad n = \frac{4}{7}.
   $$
   Получим:
   $$
   p_t = \frac{4 - 2}{\frac{3}{2}} - \frac{\frac{4}{7}}{\frac{3}{2}} p_{t-1}.
   $$

4. Упростим:
   $$
   p_t = \frac{2}{\frac{3}{2}} - \frac{\frac{4}{7}}{\frac{3}{2}} p_{t-1}.
   $$
   $$
   p_t = \frac{2 \cdot 2}{3} - \frac{4 \cdot 2}{7 \cdot 3} p_{t-1}.
   $$
   $$
   p_t = \frac{4}{3} - \frac{8}{21} p_{t-1}.
   $$

5. Это разносное уравнение первого порядка:
   $$
   p_t = \frac{4}{3} - \frac{8}{21} p_{t-1}.
   $$

6. Общий вид решения для такого уравнения:
   $$
   p_t = C \left( -\frac{8}{21} \right)^t + p_{\text{уст}},
   $$
   где $ p_{\text{уст}} $ — установившееся значение цены, которое находится из равновесного состояния. 

   Для равновесного состояния, когда $ p_t = p_{t-1} = p_{\text{уст}} $:
   $$
   p_{\text{уст}} = \frac{\frac{4}{3}}{1 + \frac{8}{21}}.
   $$

7. Упростим:
   $$
   p_{\text{уст}} = \frac{\frac{4}{3}}{\frac{21 + 8}{21}} = \frac{\frac{4}{3} \cdot 21}{29} = \frac{28}{29}.
   $$

8. Таким образом, общее решение:
   $$
   p_t = C \left( -\frac{8}{21} \right)^t + \frac{28}{29}.
   $$

Здесь $ C $ — произвольная постоянная, определяемая из начального условия $ p_0 $.
